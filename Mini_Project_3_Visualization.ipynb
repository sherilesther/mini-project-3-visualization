# Mini Project 3: Data Visualization and Descriptive Statistics

This notebook performs data visualization and statistical analysis on a cleaned sales dataset to understand distributions, trends, and outliers.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("cleaned_sales_data.csv")
df.head()
```

```python
df.describe()
```

```python
sales_mean = df["Sales"].mean()
sales_median = df["Sales"].median()
sales_std = df["Sales"].std()

sales_mean, sales_median, sales_std
```

```python
plt.figure()
plt.hist(df["Sales"], bins=10)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum()

sales_by_category.plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Quantity"], df["Sales"])
plt.title("Sales vs Quantity")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df["Sales"])
plt.title("Boxplot of Sales")
plt.show()
```

## Analysis and Interpretation

The histogram shows that most sales values are concentrated around the average range, indicating a moderately centered distribution.

The bar chart shows that some product categories contribute more to total sales than others, highlighting differences in category performance.

The scatter plot indicates a positive relationship between quantity sold and sales value, suggesting that higher quantities generally result in higher sales.

The boxplot reveals a few extreme sales values, which may represent special cases such as bulk purchases or potential data entry anomalies.

Overall, the visualizations and statistics help in understanding sales behavior and identifying key patterns in the data.
